### 1. Nan 특정값 대체
- 업체명 값값이 NaN인 경우, 그 값을 같은 행의 제조사명으로 대체
- 식품중량이 NaN인 경우, 그 값을 같은 행의 영양성분함량기준량으로 대체

```python
df['업체명'] = df['업체명'].fillna(df['제조사명'])
df['식품중량'] = df['식품중량'].fillna(df['영양성분함량기준량'])

print(df['업체명'].isnull().sum())  # 결측치 개수 확인
print(df['식품중량'].isnull().sum())  # 결측치 개수 확인
print(df.head())  # 결측치 처리 후 데이터 확인
```

### 2. 다른 값들의 대체

- 문자열일 경우 -> "" 으로
- 숫자일 경우 -> 0 으로

```python
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(0)
    else:
        df[col] = df[col].fillna('')
```

```python
print(df.isnull().sum())  # 결측치 처리 후 데이터 확인
```

### 3. 중복값 확인
- 주요 키워드 찾기... 어떻게 찾지?
- idea 1 : {주요음식} + _ => 모두 평균처리
    - 예: 국밥_소고기, 국밥_돼지, 국밥_돼지 => 전체 평균처리해서 국밥으로 일원화
- idea 2 : 가능한 것만 골라내서 분리하기
    - 예: 국밥_소고기, 국밥_콩나물 => 국밥과 콩나물국밥으로 분리

```python
# 언더바("_") 없앤 열 추가
df["메뉴_기본"] = df["식품명"].str.split("_").str[0]
```

```python
# 수치형, 문자형 col 구분
numeric_cols = df.select_dtypes(include='number').columns.tolist()
# 문자형은 식품명과 그룹키(메뉴_기본) 제외
string_cols = [c for c in df.select_dtypes(include='object').columns
               if c not in ['식품명', '메뉴_기본']]
```

```python
# 그룹별 aggreation 함수 정의
def mean_except_zero(x):
    # 0을 제외한 평균, 모두 0이면 0 반환
    return x[x != 0].mean() if len(x[x != 0]) > 0 else 0


agg_dict = {col: mean_except_zero for col in numeric_cols}
agg_dict.update({col: 'first' for col in string_cols})
```

```python
# 그룹화, 집계
df_grouped = df.groupby('메뉴_기본').agg(agg_dict)
```
