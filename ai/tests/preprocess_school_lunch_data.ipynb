# 1. 결측치 처리

```python
# 기본 데이터 이해
print("데이터의 기본 정보")
print(df.info())
# 데이터의 행과 열 개수
print("데이터의 행과 열 개수")
print(df.shape)
# 데이터의 통계적 요약
print("데이터의 통계적 요약")
print(df.describe())
# 데이터의 열 이름
print("데이터의 열 이름")
print(df.columns)
# 데이터의 결측치 확인
print("데이터의 결측치 확인")
print(df.isnull().sum())
# 데이터의 중복 확인
print("데이터의 중복 확인")
print(df.duplicated().sum())
# 데이터의 고유값 확인
print("데이터의 고유값 확인")
print(df.nunique())
```

### 1. Nan 특정값 대체
- 업체명 값값이 NaN인 경우, 그 값을 같은 행의 제조사명으로 대체
- 식품중량이 NaN인 경우, 그 값을 같은 행의 영양성분함량기준량으로 대체

```python
df['업체명'] = df['업체명'].fillna(df['제조사명'])
df['식품중량'] = df['식품중량'].fillna(df['영양성분함량기준량'])

print(df['업체명'].isnull().sum())  # 결측치 개수 확인
print(df['식품중량'].isnull().sum())  # 결측치 개수 확인
print(df.head())  # 결측치 처리 후 데이터 확인
```

### 2. 다른 값들의 대체

- 문자열일 경우 -> "" 으로
- 숫자일 경우 -> 0 으로

```python
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(0)
    else:
        df[col] = df[col].fillna('')
```

```python
print(df.isnull().sum())  # 결측치 처리 후 데이터 확인
```

# 2. 업체명 해당 없음 선택

```python
# 업체명 해당 없음 추려내기
df = df[df['업체명'] == '해당없음']
print(df.head())  # 업체명 해당 없음 데이터 확인
```
